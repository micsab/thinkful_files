```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
champ = pd.read_csv('nba_finals_data/champs_series_averages.csv')
runner = pd.read_csv('nba_finals_data/runner_ups_series_averages.csv')
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x=champ['Year'], y=champ['PTS'])
plt.title('Points Per Game Averaged by Champs in NBA Finals')
plt.show()
```

The above scatterplot shows the average points per game scored by the NBA Finals champion going back to 1980.  Interestingly, average points scored has been lower since the mid-90's (save for the most recent year).  NBA games today are played at a faster pace, so I would have assumed that teams would be scoring more.  Maybe people just didn't play defense back in the 80's.

```python
fig = plt.figure(figsize=(20, 5))
x1 = champ['Year']
y1 = champ['PTS']

x2 = runner['Year']
y2 = runner['PTS']

plt.subplot(1, 2, 1)
plt.plot(x1, y1, color='green', label='test')
plt.plot(x2, y2, color='red', label='line 2')
plt.title('Average Points Per Game Champs vs Runner-Ups')

plt.legend(['Champ Avg PPG', 'Runner-Up Avg PPG'])

plt.show()
```

Decided to plot average points per game between Finals champions and runner-ups.  As expected, champs come out ahead to varying degrees, although there are a few years where the runner-ups actually averaged more points, undoubtably due to 1 or 2 outlier games where they scored way more points than the champs.  Not much consolation for losing though.

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(champ['Year'], champ['STL'], color='red')

plt.subplot(1, 2, 2)
plt.plot(champ['Year'], champ['BLK'], color='blue')

plt.show()
```

```python
b = champ['FGP']
r = runner['FGP']
plt.hist(b, alpha=.5)
plt.hist(r, alpha=.5)
plt.legend(['Champ Avg FGP', 'Runner-Up Avg FGP'])
plt.title('Average Field Goal Percentage')
plt.show()
```

Plotted field goal percentage between champs and runner-ups as a hitogram.  As expected, champs have a distribution that is slightly skewed right.  While the runner-ups (losers) is bell-shaped/skewed slightly left.

```python
data = [b, r]
labels = ['Champ', 'Runner-Up']
# Question for mentor, why is this fig1 piece needed?
fig1, ax1 = plt.subplots()
ax1.boxplot(data, labels=labels)
ax1.set_title('Field Goal Percentage Distribution')
plt.show()
```

Field goal percentage distribution between champs and runner-ups.  Not unexpected.  The team that loses tends to, on average, shoot slightly worse.
