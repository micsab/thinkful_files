```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
raw_data = pd.read_csv('epi_r.csv')
```

```python
list(raw_data.columns)
```

```python
raw_data.rating.describe()
```

```python
raw_data.rating.hist(bins=20)
plt.title('Histogram of Recipe Ratings')
plt.show()
```

```python
null_count = raw_data.isnull().sum()
null_count[null_count>0]
```

```python
from sklearn.svm import SVR
svr = SVR()
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1)
Y = raw_data.rating
svr.fit(X,Y)
```

```python
plt.scatter(Y, svr.predict(X))
```

```python
svr.score(X, Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svr, X, Y, cv=5)
```

```python
recipes = pd.read_csv('epi_r.csv')
recipes['high_rating'] = np.where(recipes['rating'] >= 4.375, 1, 0)
recipes[['rating', 'high_rating']].head()
```

```python
from sklearn.linear_model import Lasso
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X = (recipes.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1) - recipes.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1).min()) / (recipes.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1).max() - recipes.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1).min())
Y = recipes['rating']

lasso = Lasso(alpha=.05)
lasso.fit(X, Y)
print(lasso.coef_)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LassoCV
clf = LassoCV(cv=5)
sfm = SelectFromModel(clf, threshold=0.1) 
sfm.fit(X, Y) 
n_features = sfm.transform(X).shape[1]
```

```python
print(n_features)
```

```python
X_transform = sfm.transform(X)
print(X_transform)
```

```python
idxs_selected = sfm.get_support(indices=True)
idxs_selected
new_features_df = X.iloc[idxs_selected]
new_features_df
```
