```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
```

```python
arrivals = pd.read_csv('airline_arrivals_2008.csv')
arrivals['CarrierDelay'].fillna(0, inplace=True)
arrivals['WeatherDelay'].fillna(0, inplace=True)
arrivals['NASDelay'].fillna(0, inplace=True)
arrivals['SecurityDelay'].fillna(0, inplace=True)
arrivals['LateAircraftDelay'].fillna(0, inplace=True)
arrivals['too_late'] = np.where(arrivals['ArrDelay'] >= 30.0, 1, 0)
arrivals.head(10)
```
