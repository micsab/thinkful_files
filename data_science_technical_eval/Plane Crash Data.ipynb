```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plane_crash = pd.read_csv('plane_crashes_data.csv')
plane_crash.head(10)
```

```python
plt.plot(plane_crash.year.value_counts().sort_index())
plt.show()
```

```python
plane_crash.groupby('year').aggregate(np.sum)
```

```python
fatal = plane_crash.groupby('year').aggregate(np.sum)
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(fatal['fatalities'])
```

```python
grouped = plane_crash.groupby('year').aggregate(np.sum)
plt.plot(grouped['fatalities']/grouped['aboard'])
plt.show()
```
