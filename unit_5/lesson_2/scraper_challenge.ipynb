```python
import os
import scrapy
import json
from scrapy.crawler import CrawlerProcess

os.remove('swfilms.json')

class MyItem(scrapy.Item):
    movie_title = scrapy.Field()

class StarSpider(scrapy.Spider):
    name = "SS"
    
    # Here is where we insert our API call.
    start_urls = [
        'https://swapi.co/api/films/?format=json'
    ]
    

    def parse(self, response):
        jsonresponse = json.loads(response.text)
        #print(jsonresponse)
        #item['movie_title'] = jsonresponse['results']
        for item in jsonresponse['results']:
            yield item
                
        #item["movie_title"] = jsonresponse["results"][2]["title"]
        
        #return item
    
process = CrawlerProcess({
    'FEED_FORMAT': 'json',
    'FEED_URI': 'swfilms.json',
    'USER_AGENT': 'ThinkfulDataScienceBootcampCrawler (thinkful.com)',
    'AUTOTHROTTLE_ENABLED': True,
    'HTTPCACHE_ENABLED': True,
})
                                         

# Starting the crawler with our spider.
process.crawl(StarSpider)
process.start()
print('Help!')
```
