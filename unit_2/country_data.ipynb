```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('countries of the world.csv', decimal=',')
df['Country'] = df.Country.str.strip()
df.head(5)
```

```python
corr = df.corr()
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(corr, vmax=.8, square=True)
plt.show()
```

```python
df['Climate Type'] = 'insufficient data'
df.loc[(df['Climate'] == 1.0), 'Climate Type'] = '1. Dry Tropical or Tundra and Ice'
df.loc[(df['Climate'] == 1.5), 'Climate Type'] = 'Both 1 & 2'
df.loc[(df['Climate'] == 2.0), 'Climate Type'] = '2. Wet Tropical'
df.loc[(df['Climate'] == 2.5), 'Climate Type'] = 'Both 2 & 3'
df.loc[(df['Climate'] == 3.0), 'Climate Type'] = '3. Temperate Humid Subtropical & Temperate Continental'
df.loc[(df['Climate'] == 4.0), 'Climate Type'] = '4. Dry Hot Summers & Wet Winters'
```

```python
df_sub = df[['Country', 'Region', 'Climate', 'Climate Type', 'Infant mortality (per 1000 births)', 
             'GDP ($ per capita)', 'Agriculture', 'Birthrate']]
df_sub.head(10)
```

```python
df_sub.corr()
```

```python
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corr, vmax=.8, square=True)
plt.show()
```

```python
df_sub.describe()
```

```python
g = sns.PairGrid(df_sub.dropna(), diag_sharey=False)
g.map_upper(plt.scatter, alpha=.5)
g.map_lower(sns.regplot, scatter_kws=dict(alpha=0))
g.map_diag(sns.kdeplot, lw=3)
plt.show()
```

```python
df_box = df_sub[['Climate Type','Infant mortality (per 1000 births)', 'GDP ($ per capita)', 
                 'Agriculture', 'Birthrate']]
df_box = pd.melt(df_box, id_vars=['Climate Type'])

g = sns.FacetGrid(df_box, col='variable', height=4, aspect=.5)
g = g.map(sns.boxplot, 'Climate Type', 'value')
plt.show()
```
