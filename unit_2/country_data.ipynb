```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('countries of the world.csv', decimal=',')
df['Country'] = df.Country.str.strip()
df.head(5)
```

```python
df['Climate Type'] = 'insufficient data'
df.loc[(df['Climate'] == 1.0), 'Climate Type'] = 'Dry Tropical or Tundra and Ice'
df.loc[(df['Climate'] == 1.5), 'Climate Type'] = 'Both 1 & 2'
df.loc[(df['Climate'] == 2.0), 'Climate Type'] = 'Wet Tropical'
df.loc[(df['Climate'] == 2.5), 'Climate Type'] = 'Both 2 & 3'
df.loc[(df['Climate'] == 3.0), 'Climate Type'] = 'Temperate Humid Subtropical & Temperate Continental'
df.loc[(df['Climate'] == 4.0), 'Climate Type'] = 'Dry Hot Summers & Wet Winters'
```

```python
df.drop(['Phones (per 1000)'], axis=1)
df.drop(['Coastline (coast/area ratio)'], axis=1)
df.drop(['Area (sq. mi.)'], axis=1)
```

```python
corr = df.corr()
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(corr, vmax=.8, square=True)
plt.show()
```
