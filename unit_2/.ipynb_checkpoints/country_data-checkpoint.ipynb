```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('countries of the world.csv', decimal=',')
df['Country'] = df.Country.str.strip()
df.head(5)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=False, figsize=(15,5))
sns.distplot(df['Population'], ax=ax1)
sns.distplot(df['Area (sq. mi.)'].dropna(), ax=ax2)
sns.distplot(df['Pop. Density (per sq. mi.)'].dropna(), ax=ax3)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=False, figsize=(15,5))
sns.distplot(df['Coastline (coast/area ratio)'].dropna(), ax=ax1)
sns.distplot(df['Net migration'].dropna(), ax=ax2)
sns.distplot(df['Infant mortality (per 1000 births)'].dropna(), ax=ax3)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=False, figsize=(15,5))
sns.distplot(df['GDP ($ per capita)'].dropna(), ax=ax1)
sns.distplot(df['Literacy (%)'].dropna(), ax=ax2)
sns.distplot(df['Phones (per 1000)'].dropna(), ax=ax3)
```

```python
g = sns.PairGrid(df.dropna(), diag_sharey=False)
g.map_upper(plt.scatter, alpha=.5)
g.map_lower(sns.regplot, scatter_kws=dict(alpha=0))
g.map_diag(sns.kdeplot, lw=3)
plt.show()
```

```python
corr = df.corr()
print(corr)
```

```python
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(corr, vmax=.8, square=True)
plt.show()
```

```python
countries_grouped= df.groupby('Region')[['Country','Population','Area (sq. mi.)']].agg({'Country':'count', 'Population':'sum','Area (sq. mi.)':'sum'})
print(countries_grouped)
```
