```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
betavar = np.random.beta(5, 1, 100)
mean = betavar.mean()
std = betavar.std()
print('Mean is {}'.format(mean))
print('Standard Deviation is {}'.format(std))
plt.hist(betavar)
plt.axvline(betavar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(betavar.mean() - betavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(betavar.mean() + betavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
paretovar = np.random.pareto(3.0, 100)
mean = paretovar.mean()
std = paretovar.std()
print('Mean is {}'.format(mean))
print('Standard Deviation is {}'.format(std))
plt.hist(paretovar)
plt.axvline(paretovar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(paretovar.mean() - paretovar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(paretovar.mean() + paretovar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
trivar = np.random.triangular(3.0, 10, 15, 100)
mean = trivar.mean()
std = trivar.std()
print('Mean is {}'.format(mean))
print('Standard Deviation is {}'.format(std))
plt.hist(trivar, density = True)
plt.axvline(trivar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(trivar.mean() - trivar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(trivar.mean() + trivar.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

```python
var1 = np.random.normal(5, 0.5, 100)
var2 = np.random.normal(10, 1, 100)
var3 = var1 + var2
mean = var3.mean()
std = var3.std()
print('Mean is {}'.format(mean))
print('Standard Deviation is {}'.format(std))
plt.hist(var3)
plt.axvline(var3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(var3.mean() - var3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(var3.mean() + var3.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```
